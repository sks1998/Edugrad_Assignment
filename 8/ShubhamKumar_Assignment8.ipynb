# Question 2

```python
file = open(r'C:\Users\SHUBHAM\Downloads\findata.csv')
data = file.read()
raw = data.split('\n')
w = raw[1:]
Open=[]
close = []
high = []
low = []
for each in w:
    if each!='':
        lis = each.split(',')
        Open.append(lis[1])
        high.append(lis[2])
        low.append(lis[3])
        close.append(lis[4])
fig1,axes1 = plt.subplots(2,2,figsize=(12,12))
axes1[0][0].plot(Open)
axes1[0][1].plot(high)
axes1[1][0].plot(low)
axes1[1][1].plot(close)
fig1.tight_layout()
```

```python
x1 = [2, 3, 5, 6, 8]
y1 = [1, 5, 10, 18, 20]
x2 = [3, 4, 6, 7, 9]
y2 = [2, 6, 11, 20, 22]
anno_opts = dict(xy=(0.5, 0.5), xycoords='axes fraction',
                 va='center', ha='center')
fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)
ax1.annotate('Figure 1', **anno_opts)
ax2 = plt.subplot2grid((3,3), (1,0))
ax3 = plt.subplot2grid((3,3), (1,1))
ax2.annotate('Figure 2', **anno_opts)
ax3.annotate('Figure 3', **anno_opts)
ax1.plot(x1,y1,color="blue",marker="*")
ax1.plot(x2,y2,color="red",marker="o")
ax2.plot(x1,y1)
ax3.plot(x2,y2)
fig.tight_layout()
```

# Question 4

```python
Popularity=[ 22.2, 17.6, 8.8, 8, 7.7, 6.7]
Programming_languages= ['Java',' Python', 'PHP', 'JavaScript', 'C#',' C++']

fig, axes = plt.subplots(2,1, figsize=(10,10))
axes[0].bar(Programming_languages, Popularity,align="center", width=0.5, alpha=0.5,color='green')
axes[0].set_xlabel('Languages')
axes[0].set_ylabel('Popularity')
axes[0].set_title('Popularity of Various Programming Languages')
axes[0].grid(which='major', axis='both',color='r', linewidth=1)

axes[1].barh(Programming_languages, Popularity,align="center",  alpha=0.5,color=('green','blue','red','orange','yellow','pink'))
axes[1].set_ylabel('Languages')
axes[1].set_xlabel('Popularity')
axes[1].set_title('Popularity of Various Programming Languages')
axes[1].grid(which='major', axis='both',color='r', linewidth=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
fig,ax = plt.subplots()
ax.set_aspect(aspect = 'equal')
plt.xlim([0,15])
plt.ylim([0,15])
C1 = mpatches.Circle((5,5),2,hatch='/',fill=False)
C2 = mpatches.Circle((10,10),1.5,fill=True, color='yellow')
ax.add_patch(C1)
ax.add_patch(C2)
plt.plot(5,5,'rx')
plt.plot(10,10,'rx')
plt.annotate('(5,5)',(5,5))
plt.annotate('(10,10)',(10,10))
plt.show()
```
